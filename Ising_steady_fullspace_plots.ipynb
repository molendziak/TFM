```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ising_1d_recon_v2.csv', sep=';')
df = df.astype(str)
for column in df.columns:
    try:
        df[column] = df[column].str.replace(',', '.').astype(float)
    except ValueError:
        pass
```

```python
x_axis = df['g/gamma'].values
Sx_exact = df['qutip_sx'].values
Sx_calc = df['Sx_calc'].values
plt.figure(figsize=(8, 6))

plt.plot(x_axis, Sx_exact, label='Exact')
plt.scatter(x_axis, Sx_calc, label='Neural Ansatz', color='darkorange')
plt.legend()
plt.grid()
plt.ylim(0, 0.6)
plt.xlabel(r'g / $\gamma$', fontsize=15)
plt.ylabel(r'$\langle Sx \rangle$', fontsize=15)
plt.show()
```

```python
Sy_exact = df['qutip_sy'].values
Sy_calc = df['Sy_calc'].values
plt.figure(figsize=(8, 6))

plt.plot(x_axis, Sy_exact, label='Exact')
plt.scatter(x_axis, Sy_calc, label='Neural Ansatz', color='darkorange')
plt.legend()
plt.xlabel(r'g / $\gamma$', fontsize=15)
plt.ylim(0, 0.5)
plt.ylabel(r'$\langle Sy \rangle$', fontsize=15)
plt.grid()
plt.show()
```

```python
Sz_exact = df['qutip_sz'].values
Sz_calc = df['Sz_calc'].values
plt.figure(figsize=(8, 6))

plt.plot(x_axis, Sz_exact, label='Exact')
plt.scatter(x_axis, Sz_calc, label='Neural Ansatz', color='darkorange')
plt.legend()
plt.grid()
plt.ylim(-1, 0)
plt.xlabel(r'g / $\gamma$', fontsize=15)
plt.ylabel(r'$\langle Sz \rangle$', fontsize=15)
plt.show()
```
